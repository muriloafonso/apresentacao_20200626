```python
### Aqui, estamos pegando o que vamos usar das bibliotecas 

### Usado para trabalhar com os dados
import pandas as pd

### Usado para dividir os dados em um conjunto de treino e outro de teste
from sklearn.model_selection import train_test_split

### Métricas 
from sklearn import metrics

### Modelos
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
### Agora, vamos pegar os dados que queremos usar

dados = pd.read_csv('dados_limpos.csv')
```

```python
# Separando os dados em um conjunto de treino e de teste

X_treino, X_teste, Y_treino, Y_teste = train_test_split(dados.iloc[:,0:8], dados['diabetes'], test_size = 0.2)
```

```python
### Chamo o modelo Floresta Aleatória
modelo1 = RandomForestClassifier(n_estimators=10)

### Treino o modelo
modelo1.fit(X_treino, Y_treino)
```

```python
### Ver a assertividade do modelo
metrics.accuracy_score(Y_teste, modelo1.predict(X_teste))
```

```python
### Chamando o modelo Regressão Logística
modelo2 = LogisticRegression(C=10)
```

```python
### Treino o modelo
modelo2.fit(X_treino, Y_treino)
```

```python
### Ver a assertividade do modelo
metrics.accuracy_score(Y_teste, modelo2.predict(X_teste))
```

```python
# No modelo 1, foi possível obter acerto de 77% sobre os dados de teste
# No modelo 2, foi possível obter acerto de 75% sobre os dados de teste
```
